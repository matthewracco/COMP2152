```python
"""
Author: Matthew Racco
Assignment: #1
"""
gym_member = "Alex Alliton" # string
preferred_weight_kg = 20.5 # float
highest_reps = 25 # int
membership_active = True # boolean
    
# keys (names) are strings, values are integers 

workout_stats = {
    "Alex": (30, 45, 20),
    "Steve": (25, 50, 10),
    "Amy": (35, 30, 25)
}

totals = {f"{friend}_Total": sum(minutes) for friend, minutes in workout_stats.items()}
workout_stats.update(totals)

workout_list = [
    list(minutes)
    for minutes in workout_stats.values()
    if isinstance(minutes, tuple)
]

yoga_running = [row[:2] for row in workout_list]
print("Yoga and Running minutes for all friends:", yoga_running)

if len(workout_list) >= 2:
    weightlifting_last_two = [row[2] for row in workout_list[-2:]]
    print("Weightlifting minutes for last two friends:", weightlifting_last_two)

for friend, minutes in workout_stats.items():
    if "_Total" in friend and minutes >= 120:
        name = friend.replace("_Total", "")
        print(f"Great job staying active, {name}!")

friend_name = input("Enter a friend's name: ")

if friend_name in workout_stats and not friend_name.endswith("_Total"):
    yoga, running, weightlifting = workout_stats[friend_name]
    total_minutes = workout_stats[f"{friend_name}_Total"]

    print(f"\n{friend_name}'s Workout Stats:")
    print(f"Yoga: {yoga} minutes")
    print(f"Running: {running} minutes")
    print(f"Weightlifting: {weightlifting} minutes")
    print(f"Total Workout Minutes: {total_minutes} minutes")
else:
    print(f"Friend {friend_name} is not found in the records.")

friends_total_minutes = {
    friend.replace("_Total", ""): minutes
    for friend, minutes in workout_stats.items()
    if friend.endswith("_Total")
}

if friends_total_minutes:
    max_friend = max(friends_total_minutes, key=friends_total_minutes.get)
    min_friend = min(friends_total_minutes, key=friends_total_minutes.get)

    print(f"\nFriend with highest total workout minutes: {max_friend} ({friends_total_minutes[max_friend]} minutes)")
    print(f"Friend with lowest total workout minutes: {min_friend} ({friends_total_minutes[min_friend]} minutes)")
else:
    print("\nNo total workout minutes found in records.")
```
